#feature selection

lean various feature selection technique

```python
from sklearn.model_selection import train_test_split
```

convert to categorical value

```python
diag_map = {'M' : 1 , 'B' : 0}
df['M'] = df['M'].map(diag_map)
```

```python
#X = df.iloc[: , 1:-1]
X = df.loc[: , df.columns!='M']
```

#Spilit into training and testing datasets

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)
```

#Model - K Fold Cross Validation

every time when we are shuffling the X_train and Y_train we are getting seprate accuracy so to avoid this we are using K Cross validation it will give us the mean of the accuracy's

```python
from sklearn.model_selection import cross_val_score
```

below we are creating two variables first one will store all the accuracy and the second will store the cross validation accuracy

```python
accuracy_all = []
cross_val_accuracy_all = []
```

the below function will take 3 parameters (model_name , X , Y , no of times we want to train our model )

#KNN

```python
scores = cross_val_score(knn,X,Y,cv=10)
accuracy_all.append(accuracy_score(knn_Y_pred,Y_test))
cross_val_accuracy_all.append(np.mean(scores))

print("Accuracy : {0:.2%}".format(accuracy_score(knn_Y_pred, Y_test)))
print("10 fold Cross Validation Score : {0:.2%} (+/-  {1:.2%})".format(np.mean(scores),np.std(scores)*2))
```
